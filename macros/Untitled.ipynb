```python
import math as ma
import numpy as np
import scipy as sc
import scipy.stats as st
import matplotlib as mpl
import matplotlib.pyplot as plt
```

```python
def my_func(x, par):
    return par[0]*x**2+par[1]*x+par[2]
def chisq(y, yfit, sigma):
    return sum(((np.asarray(y)-np.asarray(yfit))/sigma)**2)
```

```python
par = [2,3,1]
y = [my_func(i,par)*(1+0.1*np.random.normal()) for i in range(100)]
x = [i for i in range(100)]
sigma = [0.1*i for i in range(100)]
sigma[0] = 1
#print(sigma)
guess = [1.8,2.9,0.8]
yfit = [my_func(i,guess) for i in range(100)]
chisquare = chisq(y,yfit, sigma)
a = []
b = []
c = []
ch = []
xx = []
best = []
bestchsq = chisq(y,yfit, sigma)
for i in range(20000):
    pnew = np.asarray(guess) + np.random.normal(0,0.001,3)
    ynew = [my_func(i,pnew) for i in range(100)]
    chsqnew = chisq(y, ynew, sigma)
    uniform = np.random.uniform()
    if( uniform < chisquare/chsqnew):
        guess = pnew
        print(i," ",uniform," ",chisquare," ",chsqnew," ",guess)
        chisquare = chsqnew
    if(bestchsq>chsqnew):
        bestchsq = chsqnew
        best = pnew.copy()
    a.append(guess[0])  
    b.append(guess[1])
    c.append(guess[2])
    xx.append(i)

ynew = [my_func(i,best) for i in range(100)]
print(par, " ", best)
plt.plot(x,y,'.')
plt.plot(x,ynew,'-')
plt.show()
plt.plot(xx,a)
plt.show()
plt.plot(xx,b)
plt.show()
plt.plot(xx,c)
plt.show()
```
