```python
import sqlite3
import matplotlib.pyplot as plt
from matplotlib import style

def plot_from_DB(query):
    c.execute(query)
    #print(c.fetchall())
    c.execute(query)
    data = c.fetchall()
    x = []
    y = []

    for row in data:
        x.append(row[0])
        y.append(row[1])
    
    plt.plot(x, y,'.')
    plt.show()
    return
```

```python
conn = sqlite3.connect("PROSPECT_CalDB.sql")
c = conn.cursor()
query = "SELECT calgroup_id, lgain_0, lgain_1 FROM segment_response WHERE seg=110"

plot_from_DB(query)
```

```python
id = 9898
query = "select x,y from named_object, graph_points where graph_points.object_id=named_object.object_id AND named_object.object_id=%d" %id
plot_from_DB(query)
from matplotlib import pylab
import plotly.plotly as py
import numpy as np
from numpy import arange,array,ones, asarray
from scipy import stats, optimize
c.execute(query)
data = c.fetchall()
x = []
y = []
for row in data:
    x.append(row[0])
    y.append(row[1])
x=np.asarray(x)
y=np.asarray(y)
# Generated linear fit
slope, intercept, r_value, p_value, std_err = stats.linregress(x,y)
line = slope*x+intercept
plt.scatter(x,y,label='Data')
plt.plot(x,line,label='Best Fit')

plt.legend(loc='best')
plt.show()

def test_func(x, m, b):
    return m*x+b

params, params_covariance = optimize.curve_fit(test_func, x, y,
                                               p0=[-0.001, 0])

print(params)
print(params_covariance)
```
